```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Student-Employability-Datasets.xlsx')
df.head()
```

```python
df['ID'] = df.index
pd.to_numeric(df['ID'], downcast='integer')
```

```python
df.columns
```

```python
df = df[['ID', 'Name of Student', 'GENERAL APPEARANCE', 'MANNER OF SPEAKING',
       'PHYSICAL CONDITION', 'MENTAL ALERTNESS', 'SELF-CONFIDENCE',
       'ABILITY TO PRESENT IDEAS', 'COMMUNICATION SKILLS',
       'Student Performance Rating', 'CLASS']]

df.head()
```

```python
df = df.reset_index(drop = True)
```

```python
replace_dict = {
    'ID' : 'id',
    'Name of Student' : 'name_of_student',
    'GENERAL APPEARANCE' : 'general_appearance',
    'MANNER OF SPEAKING' : 'manner_of_speaking',
    'PHYSICAL CONDITION' : 'physical_condition',
    'MENTAL ALERTNESS' : 'mental_alertness',
    'SELF-CONFIDENCE' : 'self_confidence',
    'ABILITY TO PRESENT IDEAS' : 'ability_to_present_ideas',
    'COMMUNICATION SKILLS' : 'communication_skills',
    'Student Performance Rating' : 'student_performance_rating',
    'CLASS' : 'result'
}
```

```python
df.rename(replace_dict, axis=1, inplace= True) 
```

```python
df.info()
```

```python
from create_db import engine, Session, Base
```

```python
session = Session(bind = engine)
```

```python
df.to_sql(con = engine, name = 'training_set', if_exists='replace', index=False)
```

```python
session.execute('Alter Table Training_Set Modify ID INTEGER PRIMARY KEY;')
session.commit()
```

```python
# Convert an existing table to sqlalchemy table object

from sqlalchemy import Table
from sqlalchemy.ext.automap import automap_base

class TrainingSet(Base):
    __table__ = Table('training_set', Base.metadata, autoload=True, autoload_with=engine)

    def __repr__(self):
        return f'[{self.ID}, {self.name_of_student}, {self.general_appearance}, {self.manner_of_speaking}, {self.physical_condition}, {self.mental_alertness}, {self.self_confidence}, {self.ability_to_present_ideas}, {self.communication_skills}, {self.student_performance_rating}, {self.result}]'
```

```python
results = session.query(TrainingSet).filter(TrainingSet.result == 'LessEmployable').all()

for r in results:
    print(r)
```
